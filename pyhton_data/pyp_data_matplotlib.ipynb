##### **3.3 设置图例和刻度**
+ **绘制正余弦函数**

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint = True)
y = np.sin(x)
y1 = np.cos(x)
```

```python
plt.plot(x,y,label = 'sinx')
plt.plot(x,y1,label = 'cosx')
plt.legend(loc = 0)
plt.show()
```

+ **规定xy轴的范围**

```python
plt.plot(x,y,label = 'sinx')
plt.plot(x,y1,label = 'cosx')
plt.xlim(-4,4)
# plt.ylim(-1,1)
plt.legend(loc = 0)
plt.show()
```

```python
plt.plot(x,y,label = 'sinx')
plt.plot(x,y1,label = 'cosx')
plt.axis([-3.2,3.2, -1,1])
plt.legend(loc = 0)
plt.show()
```

+ **设置刻度**

```python
plt.plot(x,y,label = 'sinx')
plt.plot(x,y1,label = 'cosx')
plt.axis([-3.2,3.2, -1,1])
plt.xticks([-np.pi,0,np.pi], [r'$-\pi$', 0, r'$+\pi$'])
plt.legend(loc = 0)
plt.show()
```

##### **3.4 绘制多图**

```python
x = [1,2,3,4]
y = [2,4,3,7]
y1 = [5,8,3,9]
```

```python
# 1行两列在一张图
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.bar(x,y1)
plt.show()
```

+ **设置图表的大小**

```python
plt.figure(figsize = (8,4))
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.bar(x,y1)
plt.show()
```

+ **多图的位置与放置**

```python
plt.figure(figsize = (10,6))
plt.subplot(2,3,1)
plt.plot(x,y)
# 柱状图
plt.subplot(2,3,2)
plt.bar(x,y1)
# 横向柱状图
plt.subplot(2,3,3)
plt.barh(x,y1)
# 散点图
plt.subplot(2,3,4)
plt.scatter(x,y1)
# 箱型图
plt.subplot(2,3,5)
plt.boxplot(y1)
# 饼状图
plt.subplot(2,3,6)
plt.pie(y1)
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.subplot(2,3,1)
plt.pie(y)
# 柱状图
plt.subplot(2,3,2)
plt.bar(x,y1)
# 横向柱状图
plt.subplot(2,3,3)
plt.barh(x,y1)
# 散点图
plt.subplot(2,1,2)
plt.plot(x,y1)
plt.show()
```

##### **3.5 箱型图、直方图和饼图**
+ **箱型图**

```python
data = [23,22,23,45,22,34,23,26,24,25,28,36,34,23,21,26,39,42,37,27,20,18]
```

```python
plt.figure(figsize = (6,4))
plt.subplot(1,1,1)
plt.boxplot(data)
plt.show()
```

+ **直方图**

```python
plt.hist(data)
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins = 100)
plt.show()
```

+ **饼图**

```python
data = [13,30,45,20]
labels = ['AA','BB','CC','DD']
explode = [0, 0, 0.1, 0]
plt.pie(data, labels = labels, autopct = '%1.1f%%', explode = explode, shadow = True, startangle = 90)
plt.axis('equal') # 设置图像的形状
plt.show()
```

##### **3.6 面向对象绘图**

```python
# 从1到11取30个值
x = np.random.randint(1,11,30)
y = np.random.randint(1,6,30)
y1 = y + 5
plt.figure(figsize = (8,6))
plt.scatter(x,y, label='one', color='g', s=25, marker='^')
plt.scatter(x,y1, label='two', color='r', s=25, marker='o')
x_ticks_label = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten']
plt.xticks(np.arange(1,11), x_ticks_label, rotation=-70, fontsize=14)

plt.grid(True)
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.title('i am title')
plt.legend()
plt.show()
```

```python
# 初始化绘图对象
fig,ax = plt.subplots()
ax.scatter(x,y, label='one', color='g', s=25, marker='^')
ax.scatter(x,y1, label='two', color='r', s=25, marker='o')
ax.set_xlabel('x_data')
ax.set_ylabel('y_data')
ax.set_title('i am title')
ax.grid(True)
```
